###  Fits data again but this time incorporating emf constant

```python
#now it's time to fit the data to a damped wave
#define a function. 

time = np.arange(0, cut_off_point, 0.0001)

def fit(time, A, b, omega_star_a, phi):
    beta = b/(2*m)    
    exp_term = -A*np.exp(-beta*time)
    sin = omega_star_a * np.sin(time*omega_star_a + phi)
    cos = beta* np.cos(omega_star_a*time + phi)

    return emf_constant* exp_term*(sin + cos)
    
popt, pcov = curve_fit(fit, time_clean_a, voltage_clean_a)


A = popt[0]
b = popt[1]
omega_star_a = popt[2]
phi = popt[3]
omega_star_err_a = pcov[2,2]**0.5

print("A =", popt[0], "+/-", pcov[0,0]**0.5)
print("b =", popt[1], "+/-", pcov[1,1]**0.5)
print("omega_star_a =", omega_star_a, "+/-", omega_star_err_a)
print("phi =", popt[3], "+/-", pcov[3,3]**0.5)

fig = plt.figure(figsize=(11,7))
ax = fig.add_subplot(111)
plt.plot(time, fit(time, A, b, omega_star_a, phi), color = "black", linewidth = 2)
plt.plot(time_clean_a, voltage_clean_a, marker = 'o', color = 'black', ls = 'none', markersize = 3)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel("Time / s", size = 15)
plt.ylabel("EMF / V", size = 15)
plt.tick_params(direction='in',      # I like 'in', could be 'out' or both 'inout'
                length=7,            # A reasonable length
                bottom='on',         # I want ticks on the bottom axes
                left='on',
                top='on',
                right='on')
plt.plot()
```
